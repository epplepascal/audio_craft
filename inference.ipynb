```python
from audiocraft.utils import export
from audiocraft import train
from audiocraft.models import MusicGen
import IPython.display as ipd
```

```python
# Signature of the model
sig = "677ae10e"
```

```python
xp = train.main.get_xp_from_sig(sig)
name_of_model = 'MY_AUDIO_LM'
export.export_lm(xp.folder / 'checkpoint.th', f'checkpoints/{name_of_model}/state_dict.bin')
export.export_pretrained_compression_model('facebook/encodec_32khz', f'checkpoints/{name_of_model}/compression_state_dict.bin')
```

```python
# load your finetune
musicgen = MusicGen.get_pretrained('checkpoints/my_audio_lm')
musicgen.set_generation_params(duration=16)
```

```python
wavs = musicgen.generate(['A cool song from Voyager.'])
generated_music = wavs[0]
```

```python
ipd.display(ipd.Audio(generated_music.cpu(), rate=32000))
```
